```python
nba_pointguards = ["Irving", "Paul", "Curry", "Mitchel", "James", "Jordan", "Barnes", "Antetokounmpo"]
nba_shootingguards = ["James", "Mitchel", "Jordan", "Cousins", "Miller", "Harden", "Bryant", "Thompson"]

pg = set (nba_pointguards)
sg = set (nba_shootingguards)
InnerJoin = pg.intersection(sg)
InnerJoin
```

```python
pg_ljoin = InnerJoin.union(pg)
pg_ljoin
```

```python
sg_rjoin = InnerJoin.union(sg)
sg_rjoin
```

```python
pg_Lexlusivejoin = pg.difference(sg)
pg_Lexlusivejoin
```

```python
sg_Rexlusivejoin = sg.difference(pg)
sg_Rexlusivejoin
```

```python
Full_OuterJoin = pg.union(sg)
Full_OuterJoin
```

```python
Full_OuterExJoin = (pg.difference(sg)).union(sg.difference(pg))
Full_OuterExJoin
```
